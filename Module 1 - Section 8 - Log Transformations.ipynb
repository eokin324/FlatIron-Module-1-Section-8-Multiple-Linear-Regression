```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("auto-mpg.csv")
df.head()
```

```python
from statsmodels.formula.api import ols
dependent_vble = 'mpg'
independent_vbles = ['displacement', 'horsepower', 'weight',
       'acceleration']
predictors = '+'.join(independent_vbles)
formula = dependent_vble + "~" + predictors
model = ols(formula=formula, data=df).fit()
print(model.summary())
```

```python
non_normal_predictors = ['displacement', 'horsepower', 'weight']
for element in non_normal_predictors:
    df[element] = df[element].map(lambda x: np.log(x))
```
